## <strong>2. NumPy: Universal Functions (UFunc)</strong>

````python
import numpy as np
````

````python
# 역수 계산 함수
def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

# [+] 함수 테스트
values = np.random.randint(1,10, size=5)
````

````python
compute_reciprocals(values)
````

#### <strong>역수 계산: Python loop vs NumPy ufuncs</strong>
+ ```%timeit```: 셀 단위 코드 실행 시간을 측정하는 매직 함수

````python
# [+] 백만 개 크기의 난수 배열 생성: 정수, 값 범위 [1, 100)
big_array = np.random.randint(1,100, size =1000000)
````

````python
big_array[:100]
````

````python
%%timeit

# [+] Python loop를 통한 역수 계산 
compute_reciprocals(big_array)
````

````python
%%timeit

# [+] NumPy ufunc을 통한 벡터화된 역수 계산
1 / big_array
````

#### <strong>유니버설 함수들</strong>

````python
# 난수 배열 2개 생성
np.random.seed(1)
arr1 = np.random.randint(10, size=5)
arr2 = np.random.randint(10, size=5)

print(arr1, arr2)
````

````python
# 배열-스칼라 연산
arr1 = arr1 - 2
print(arr1)
````

````python
# 배열-배열 연산
arr3 = arr1 + arr2
print(arr3)
````

````python
# 다차원 배열 연산: 2차원 배열 ** 스칼라  차원이 다른 배열끼리 연산을 빠르게 가능케 함 - numpy
arr4 = np.random.randint(10, size=(3, 5))
print(arr4)

arr5 = arr4 ** 2
print(arr5)
````

````python
"""
    UFuncs: 산술 연산
"""

# 기본 산술 연산
x = np.arange(4)
print("x     =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x ** 2 =", x ** 2)

# 산술 연산에 대응하는 유니버설 함수
print("x + 5 =", np.add(x, 5))
print("x - 5 =", np.subtract(x, 5))
print("x * 2 =", np.multiply(x, 2))
print("x / 2 =", np.divide(x, 2))
print("x ** 2 =", np.power(x, 2))
````

````python
"""
    UFuncs: 절댓값 계산
"""

# [+] 절댓값 함수
x = np.array([-2, -1, 0, 1, 2])
np.abs(x)
````

````python
"""
    np.linspace(start, stop, num): 선형 간격 벡터 생성
        - start: 벡터의 시작 값
        - end: 벡터의 종료 값
        - num: 생성할 값의 개수
"""

# [+][0, 100]의 범위를 가지며, 균일한 간격의 5개의 값들로 구성되는 벡터를 생성
x = np.linspace(0,100, 5)
x
````

````python
# [+] [0, 100]의 범위를 가지며, [0, 10, 20, 30, ..., 100]와
# 같은 값들로 구성되는 벡터를 생성
x = np.linspace(0,100, 11)
x
````

````python
# 삼각함수(trigonometric functions)
theta = np.linspace(0, np.pi, 3)

print(theta)
print(np.sin(theta))
print(np.cos(theta))
print(np.tan(theta))
````

````python
"""
    집계 함수: reduce()
"""

# 배열 생성
x = np.arange(1, 5)
x
````

````python
# [+] 덧셈 연산으로 집계
aggr = np.add.reduce(x)
aggr
````

````python
# [+] 곱셈 연산으로 집계
aggr = np.multiply.reduce(x)
aggr
````

````python
"""
    집계 함수: accumulate()
"""

# [+] 덧셈 연산으로 집계
accum = np.add.accumulate(x)
accum
````

````python
# [+] 곱셈 연산으로 집계
accum = np.multiply.accumulate(x)
accum
````

````python
# 집계 함수: sum()
np.random.seed(3) #랜덤 시드 for 재현성
arr =  np.random.rand(1000000)# [+] 0~1 사이의 실수 형태의 난수 배열 생성

print(arr)
print("sum =", np.sum(arr))
````

````python
# sum(): Python 내장함수
%timeit sum(arr)
````

````python
# np.sum()
%timeit np.sum(arr) # 1000µs = 1ms
````

````python
# [+] 그 외의 집계함수
print("max =", np.max(arr))
print("min =", np.min(arr))
print("mean =", np.mean(arr))
print("median =", np.median(arr))
print("percentile =", np.percentile(arr , 90))
print("variance =", np.var(arr))
print("standard deviation =", np.std(arr))
print("index of maximum value =", np.argmax(arr))
print("index of minimum value =", np.argmin(arr))
print("Is there any number greater than 1? =", np.any(arr > 1))
print("Is every number greater than 0? =", np.all(arr)) 
````
